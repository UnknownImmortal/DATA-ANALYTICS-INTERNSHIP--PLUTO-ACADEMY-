```python
import pandas as pd

customers = pd.read_csv("olist_customers_dataset.csv")

orders = pd.read_csv("olist_orders_dataset.csv")

products = pd.read_csv("olist_products_dataset.csv")

order_items = pd.read_csv("olist_order_items_dataset.csv")

reviews = pd.read_csv("olist_order_reviews_dataset.csv")

payments = pd.read_csv("olist_order_payments_dataset.csv")

sellers = pd.read_csv("olist_sellers_dataset.csv")

geolocation = pd.read_csv("olist_geolocation_dataset.csv")

translation = pd.read_csv("product_category_name_translation.csv")

print("All datasets loaded successfully!")
```

```python
orders['order_purchase_timestamp'] = pd.to_datetime(
    orders['order_purchase_timestamp']
)

orders['order_approved_at'] = pd.to_datetime(
    orders['order_approved_at']
)

orders['order_delivered_carrier_date'] = pd.to_datetime(
    orders['order_delivered_carrier_date']
)

orders['order_delivered_customer_date'] = pd.to_datetime(
    orders['order_delivered_customer_date']
)

orders['order_estimated_delivery_date'] = pd.to_datetime(
    orders['order_estimated_delivery_date']
)

print("Order dates converted successfully")
```

```python
reviews['review_creation_date'] = pd.to_datetime(
    reviews['review_creation_date']
)

reviews['review_answer_timestamp'] = pd.to_datetime(
    reviews['review_answer_timestamp']
)

print("Review dates converted successfully")
```

```python
order_items['shipping_limit_date'] = pd.to_datetime(
    order_items['shipping_limit_date']
)

print("Shipping date converted successfully")
```

```python
orders.info()
```

```python
products['product_category_name'] = products[
    'product_category_name'
].fillna('Unknown')
```

```python
reviews['review_comment_title'] = reviews[
    'review_comment_title'
].fillna('No Title')

reviews['review_comment_message'] = reviews[
    'review_comment_message'
].fillna('No Comment')

print("Missing review comments handled")
```

```python
print("Orders duplicates:", orders.duplicated().sum())

print("Products duplicates:", products.duplicated().sum())

print("Reviews duplicates:", reviews.duplicated().sum())
```

```python
df = orders.merge(
    customers,
    on='customer_id',
    how='left'
)

df = df.merge(
    order_items,
    on='order_id',
    how='left'
)

df = df.merge(
    products,
    on='product_id',
    how='left'
)

df = df.merge(
    reviews,
    on='order_id',
    how='left'
)

print(df.shape)
```

```python
df.to_csv("cleaned_ecommerce_data.csv", index=False)

print("Cleaned dataset saved successfully")
```

```python
category_sales = df.groupby(
    'product_category_name'
)['price'].sum().sort_values(ascending=False)

print(category_sales.head(10))
```

```python
top10 = category_sales.head(10)

plt.figure(figsize=(12,6))

sns.barplot(
    x=top10.values,
    y=top10.index
)

plt.title("Top 10 Product Categories by Revenue")

plt.xlabel("Revenue")

plt.ylabel("Product Category")

plt.show()
```

```python
df['month'] = df[
    'order_purchase_timestamp'
].dt.month_name()
```

```python
monthly_sales = df.groupby(
    'month'
)['price'].sum().sort_values(ascending=False)

print(monthly_sales)
```

```python
month_order = [
    'January', 'February', 'March',
    'April', 'May', 'June',
    'July', 'August', 'September',
    'October', 'November', 'December'
]

monthly_trend = df.groupby(
    'month'
)['price'].sum().reindex(month_order)

plt.figure(figsize=(12,6))

sns.lineplot(
    x=monthly_trend.index,
    y=monthly_trend.values,
    marker='o'
)

plt.title("Monthly Sales Trend")

plt.xlabel("Month")

plt.ylabel("Revenue")

plt.xticks(rotation=45)

plt.show()
```

```python
state_sales = df.groupby(
    'customer_state'
)['price'].sum().sort_values(ascending=False)

print(state_sales.head(10))
```

```python
top_states = state_sales.head(10)

plt.figure(figsize=(12,6))

sns.barplot(
    x=top_states.index,
    y=top_states.values
)

plt.title("Top States by Revenue")

plt.xlabel("State")

plt.ylabel("Revenue")

plt.show()
```

```python
avg_order_value = df.groupby(
    'month'
)['price'].mean().reindex(month_order)

print(avg_order_value)
```

```python
plt.figure(figsize=(12,6))

sns.lineplot(
    x=avg_order_value.index,
    y=avg_order_value.values,
    marker='o'
)

plt.title("Average Order Value Trend")

plt.xlabel("Month")

plt.ylabel("Average Order Value")

plt.xticks(rotation=45)

plt.show()
```

```python
review_distribution = df[
    'review_score'
].value_counts().sort_index()

print(review_distribution)
```

```python
plt.figure(figsize=(8,8))

plt.pie(
    review_distribution.values,
    labels=review_distribution.index,
    autopct='%1.1f%%'
)

plt.title("Customer Review Score Distribution")

plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.histplot(
    df['price'],
    bins=50
)

plt.title("Distribution of Order Values")

plt.xlabel("Order Value")

plt.ylabel("Frequency")

plt.show()
```

```python
heatmap_data = pd.pivot_table(
    df,
    values='price',
    index='product_category_name',
    columns='month',
    aggfunc='sum'
)

top_categories = category_sales.head(10).index

heatmap_data = heatmap_data.loc[top_categories]

heatmap_data = heatmap_data[month_order]
```

```python
plt.figure(figsize=(14,8))

sns.heatmap(
    heatmap_data,
    cmap='YlGnBu'
)

plt.title("Category vs Month Revenue Heatmap")

plt.xlabel("Month")

plt.ylabel("Category")

plt.show()
```
